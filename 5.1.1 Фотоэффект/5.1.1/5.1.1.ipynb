```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

from scipy.optimize import curve_fit

def quadratic_function(x, a, b, c):
        return a * x**2 + b * x + c

# Загрузка данных из CSV файла
data = np.genfromtxt('calibration.csv', delimiter=',', names=True)

x = data['phi']
y = data['lambda']

# Initial guesses for a, b, c (optional but recommended)
initial_guesses = [1, 1, 1]

# Perform the curve fit
params, covariance = curve_fit(quadratic_function, x, y, p0=initial_guesses)

# Extract the fitted parameters
a_fit, b_fit, c_fit = params
print(f"Fitted parameters: a={a_fit}, b={b_fit}, c={c_fit}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Загрузка данных из CSV файла
data = np.genfromtxt('sqrtV-V.csv', delimiter=',', names=True)

x = data['V']
y = data['I']

# Построение точек на графике
plt.scatter(x, y, label='R(Q), T = 296.2 K', s=10, c=(0,0,0))

# Создаем и обучаем модель линейной регрессии
new_x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(new_x, y)

# Получаем коэффициенты зависимости
a = model.coef_[0]
b = model.intercept_
print(f"k: {a}")
print(f"b: {b}")

# Предсказываем значения
y_pred = model.predict(new_x)

# Среднеквадратичная ошибка модели
mse = mean_squared_error(y, y_pred)

# Количество наблюдений
n = len(x)

# Стандартная ошибка наклона a
se_a = np.sqrt(mse / ((n - 1) * np.var(x)))

# Стандартная ошибка смещения b
se_b = np.sqrt(mse * (1/n + np.mean(x)**2 / ((n - 1) * np.var(x))))

print("Стандартная ошибка наклона k:", se_a)
print("В процентах: ", se_a / a * 100)
print("Стандартная ошибка смещения b:", se_b)
print("В процентах: ", se_b / b * 100)

# МНК для построения сглаживающей прямой
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Построение сглаживающей прямой
plt.plot(x, y_fit, color='pink')

plt.xlabel('Q')
plt.ylabel('R')
plt.legend()
plt.show()
```
