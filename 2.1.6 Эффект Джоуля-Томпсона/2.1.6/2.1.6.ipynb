```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

################################################################################
# Загрузка данных из CSV файла 20.csv
data = np.genfromtxt('20.csv', delimiter=',', names=True)


x = data['a']
y = data['e']

# Построение точек на графике
plt.scatter(x, y, label=' ', s=10, c=(0,0,0))

# Создаем и обучаем модель линейной регрессии
new_x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(new_x, y)

# Получаем коэффициенты зависимости
a = model.coef_[0]
b = model.intercept_
print(f"k: {a}")
print(f"b: {b}")

# Предсказываем значения
y_pred = model.predict(new_x)

# Среднеквадратичная ошибка модели
mse = mean_squared_error(y, y_pred)

# Количество наблюдений
n = len(x)

# Стандартная ошибка наклона a
se_a = np.sqrt(mse / ((n - 1) * np.var(x)))

# Стандартная ошибка смещения b
se_b = np.sqrt(mse * (1/n + np.mean(x)**2 / ((n - 1) * np.var(x))))

print("Стандартная ошибка наклона k:", se_a)
print("В процентах: ", se_a / a * 100)
print("Стандартная ошибка смещения b:", se_b)
print("В процентах: ", se_b / b * 100)

# МНК для построения сглаживающей прямой
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Построение сглаживающей прямой
plt.plot(x, y_fit, color='pink', label='T = 20^oC')
################################################################################
# Загрузка данных из CSV файла 30.csv
data = np.genfromtxt('30.csv', delimiter=',', names=True)


x = data['x']
y = data['y']

# Построение точек на графике
plt.scatter(x, y, label=' ', s=10, c='red')

# Создаем и обучаем модель линейной регрессии
new_x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(new_x, y)

# Получаем коэффициенты зависимости
a = model.coef_[0]
b = model.intercept_
print(f"k: {a}")
print(f"b: {b}")

# Предсказываем значения
y_pred = model.predict(new_x)

# Среднеквадратичная ошибка модели
mse = mean_squared_error(y, y_pred)

# Количество наблюдений
n = len(x)

# Стандартная ошибка наклона a
se_a = np.sqrt(mse / ((n - 1) * np.var(x)))

# Стандартная ошибка смещения b
se_b = np.sqrt(mse * (1/n + np.mean(x)**2 / ((n - 1) * np.var(x))))

print("Стандартная ошибка наклона k:", se_a)
print("В процентах: ", se_a / a * 100)
print("Стандартная ошибка смещения b:", se_b)
print("В процентах: ", se_b / b * 100)

# МНК для построения сглаживающей прямой
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Построение сглаживающей прямой
plt.plot(x, y_fit, color='purple', label='T = 30^oC')
################################################################################
# Загрузка данных из CSV файла 40.csv
data = np.genfromtxt('40.csv', delimiter=',', names=True)


x = data['x']
y = data['y']

# Построение точек на графике
plt.scatter(x, y, label=' ', s=10, c='blue')

# Создаем и обучаем модель линейной регрессии
new_x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(new_x, y)

# Получаем коэффициенты зависимости
a = model.coef_[0]
b = model.intercept_
print(f"k: {a}")
print(f"b: {b}")

# Предсказываем значения
y_pred = model.predict(new_x)

# Среднеквадратичная ошибка модели
mse = mean_squared_error(y, y_pred)

# Количество наблюдений
n = len(x)

# Стандартная ошибка наклона a
se_a = np.sqrt(mse / ((n - 1) * np.var(x)))

# Стандартная ошибка смещения b
se_b = np.sqrt(mse * (1/n + np.mean(x)**2 / ((n - 1) * np.var(x))))

print("Стандартная ошибка наклона k:", se_a)
print("В процентах: ", se_a / a * 100)
print("Стандартная ошибка смещения b:", se_b)
print("В процентах: ", se_b / b * 100)

# МНК для построения сглаживающей прямой
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Построение сглаживающей прямой
plt.plot(x, y_fit, color='pink', label='T = 40^oC')
################################################################################
# Загрузка данных из CSV файла 50.csv
data = np.genfromtxt('50.csv', delimiter=',', names=True)


x = data['x']
y = data['y']

# Построение точек на графике
plt.scatter(x, y, label=' ', s=10, c='green')

# Создаем и обучаем модель линейной регрессии
new_x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(new_x, y)

# Получаем коэффициенты зависимости
a = model.coef_[0]
b = model.intercept_
print(f"k: {a}")
print(f"b: {b}")

# Предсказываем значения
y_pred = model.predict(new_x)

# Среднеквадратичная ошибка модели
mse = mean_squared_error(y, y_pred)

# Количество наблюдений
n = len(x)

# Стандартная ошибка наклона a
se_a = np.sqrt(mse / ((n - 1) * np.var(x)))

# Стандартная ошибка смещения b
se_b = np.sqrt(mse * (1/n + np.mean(x)**2 / ((n - 1) * np.var(x))))

print("Стандартная ошибка наклона k:", se_a)
print("В процентах: ", se_a / a * 100)
print("Стандартная ошибка смещения b:", se_b)
print("В процентах: ", se_b / b * 100)

# МНК для построения сглаживающей прямой
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Построение сглаживающей прямой
plt.plot(x, y_fit, color='purple', label='T = 50^oC')
################################################################################

plt.xlabel('$\Delta P$, атм')
plt.ylabel('$\Delta T$, $^o$К')
plt.legend()
plt.show()
```
