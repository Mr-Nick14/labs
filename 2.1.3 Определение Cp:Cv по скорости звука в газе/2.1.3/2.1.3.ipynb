```python
import numpy as np
import matplotlib.pyplot as plt
import csv

# Загрузка данных из CSV файла
data = np.genfromtxt('first.csv', delimiter=',', names=True)

x = data['n']
y = data['f']

# Построение точек на графике
plt.scatter(x, y, label='f(n)', s=10, c=(0,0,0))

# МНК для построения сглаживающей прямой
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Построение сглаживающей прямой
plt.plot(x, y_fit, color='pink')

plt.xlabel('n')
plt.ylabel('f')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

# Загрузка данных из CSV файла
data = np.genfromtxt('c(T).csv', delimiter=',', names=True)

x = data['T']
y = data['c']
errors = data['error']

# Построение точек на графике
plt.errorbar(x, y, yerr=errors, fmt='o', color='b', ecolor='r', capsize=5, capthick=0.5)
plt.scatter(x, y, s=10, c=(0,0,0))


# МНК для построения сглаживающей прямой
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Построение сглаживающей прямой
plt.plot(x, y_fit, color='pink')

plt.xlabel('T')
plt.ylabel('c')
plt.legend()
plt.show()
```
